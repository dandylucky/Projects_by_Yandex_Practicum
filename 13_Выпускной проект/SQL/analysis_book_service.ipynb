### Cколько книг вышло после 1 января 2000 года

```python
task_1 = ''' SELECT COUNT(book_id)
             FROM books
             WHERE publication_date > '2000-01-01' 
        '''
```

```python
result(task_1)
```

**Ответ:**

819 книг было выпущено после 1 января 2000 года

###  Количество обзоров и средняя оценка для каждой книги

```python
task_2 = ''' SELECT b.book_id, b.title, reviews_count, avg_rating
             FROM books AS b
             LEFT JOIN (
                        SELECT rev.book_id, COUNT(rev.review_id) AS reviews_count
                        FROM reviews AS rev
                        GROUP BY rev.book_id) AS tab_1 ON b.book_id = tab_1.book_id
             LEFT JOIN (
                        SELECT rat.book_id, ROUND(AVG(rat.rating),2) AS avg_rating
                        FROM ratings AS rat
                        GROUP BY rat.book_id) AS tab_2 ON b.book_id = tab_2.book_id         
         '''
```

```python
result(task_2)
```

**Ответ:**
    
Таблица с количеством обзоров и средней оценкой для каждой книги    

### Издательство, которое выпустило наибольшее число книг толще 50 страниц

```python
task_3 = ''' SELECT p.publisher, COUNT(b.book_id) AS books_count
             FROM books AS b
             LEFT JOIN publishers AS p ON b.publisher_id = p.publisher_id
             WHERE b.num_pages > 50
             GROUP BY p.publisher
             ORDER BY books_count DESC
             LIMIT 1
        '''
```

```python
result(task_3)
```

**Ответ:**

Издательство Penguin Books выпустило 42 книги, толще 50 страниц

### Автор с самой высокой средней оценкой книг — учитывать только книги с 50 и более оценками

```python
task_4 = ''' SELECT a.author, ROUND(AVG(r.rating), 2) AS avg_rating
             FROM books AS b
             LEFT JOIN authors AS a ON b.author_id = a.author_id 
             LEFT JOIN ratings AS r ON b.book_id = r.book_id
             WHERE b.book_id IN (
                                 SELECT book_id
                                 FROM ratings 
                                 GROUP BY book_id
                                 HAVING COUNT(rating_id) > 50)
             GROUP BY a.author
             ORDER BY AVG(r.rating) DESC
             LIMIT 1
         '''
```

```python
result(task_4)
```

**Ответ:**

J.K. Rowling - автор с самой высокой средней оценкой книг (4.29)

### Cреднее количество обзоров от пользователей, которые поставили больше 50 оценок.

```python
task_5 = '''  SELECT ROUND(AVG(count_rev)) AS avg_reviews_count
              FROM (
                      SELECT username, COUNT(review_id) AS count_rev
                      FROM reviews
                      WHERE username IN (
                                          SELECT username
                                          FROM ratings 
                                          GROUP BY username
                                          HAVING COUNT(rating_id) > 50)
                     GROUP BY username) AS tab_1 
        '''
```

```python
result(task_5)
```

**Ответ:**

Пользователи, которые поставили больше 50 оценок, в среднем пишут 24 отзыва
